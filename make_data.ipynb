```python
# Split the data into lines and then split each line into parts
lines = data.split('\n')
```

```python
cleaned_lines = [line.replace('             ','  0000 0000  ') for line in lines if len(line.strip()) > 0]
```

```python
cleaned_lines
```

```python
parsed_data = [line.split() for line in cleaned_lines]
```

```python
import pandas as pd
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
structured_data = []

# Process each line of the parsed data
for line in parsed_data:
    if len(line) != 25:
        print(line)
        print("Data format error: Skipping line.")
        continue
    day = line[0]
    for i in range(12):
        month = months[i]
        rise = line[1 + 2 * i]
        set_ = line[2 + 2 * i]
        structured_data.append({'month': f'{month}-{day}-2024', 'rise': rise, 'set': set_})

# Create the DataFrame
df = pd.DataFrame(structured_data)

print(df.head())
```

```python
# parsed_data
```

```python
df = df[df['rise'] != '0000']
```

```python
df.head()
```

```python
from datetime import datetime
def convert_time(time_str):
    time_obj = datetime.strptime(time_str, '%H%M')
    return time_obj.strftime('%I:%M %p')

df['rise'] = df['rise'].apply(convert_time)
df['set'] = df['set'].apply(convert_time)
```

```python
df['month'] = pd.to_datetime(df['month'], format='%b-%d-%Y')
```

```python
df.to_csv('ChicagoSunriseSet2024.csv', index=False)
```
